```python
#LIMPIEZA
for columna in VIGILANCIA.columns[:] :
   VIGILANCIA['Índice de Breteau'] = VIGILANCIA['Índice de Breteau'].astype(str).str.replace('#DIV/0!', '0')
   VIGILANCIA['Índice de Recipiente'] = VIGILANCIA['Índice de Recipiente'].astype(str).str.replace('#DIV/0!', '0')
   VIGILANCIA['Índice de viviendas/inmuebles'] = VIGILANCIA['Índice de viviendas/inmuebles '].astype(str).str.replace('#DIV/0!', '0')
   VIGILANCIA['Cantidad total de PREDIOS ESTRATÉGICOS visitados'] = VIGILANCIA['Cantidad total de PREDIOS ESTRATÉGICOS visitados'].astype(str).str.replace('O', '0')
   VIGILANCIA['Cantidad total de PREDIOS ESTRATÉGICOS visitados'] = VIGILANCIA['Cantidad total de PREDIOS ESTRATÉGICOS visitados'].astype(str).str.replace('|0', '0')
   VIGILANCIA['Cantidad total de VIVIENDAS visitadas'] = VIGILANCIA['Cantidad total de VIVIENDAS visitadas'].astype(str).str.replace('O', '0')
   VIGILANCIA['Cantidad total de VIVIENDAS visitadas'] = VIGILANCIA['Cantidad total de VIVIENDAS visitadas'].astype(str).str.replace('p', '0')
   VIGILANCIA['Cantidad total de VIVIENDAS visitadas'] = VIGILANCIA['Cantidad total de VIVIENDAS visitadas'].astype(str).str.replace('+', '0')
   VIGILANCIA['Cantidad total de VIVIENDAS visitadas'] = VIGILANCIA['Cantidad total de VIVIENDAS visitadas'].astype(str).str.replace('9', '0')
   VIGILANCIA['Cantidad total de VIVIENDAS visitadas'] = VIGILANCIA['Cantidad total de VIVIENDAS visitadas'].astype(str).str.replace('6', '0')
   VIGILANCIA['Cantidad de inmuebles INSPECCIONADOS'] = VIGILANCIA['Cantidad de inmuebles INSPECCIONADOS'].astype(str).str.replace('O', '0')
   VIGILANCIA['Total de inmuebles positivos'] = VIGILANCIA['Total de inmuebles positivos'].astype(str).str.replace('O', '0')
   VIGILANCIA['Recipientes POSITIVOS'] = VIGILANCIA['Recipientes POSITIVOS'].astype(str).str.replace('O', '0')
   VIGILANCIA['Recipientes NEGATIVOS'] = VIGILANCIA['Recipientes NEGATIVOS'].astype(str).str.replace('O', '0')
   VIGILANCIA['Total de ELIMINADOS'] = VIGILANCIA['Total de ELIMINADOS'].astype(str).str.replace('O', '0')
```

```python
# Primero seleccionamos las columnas que queremos "encolumnar"

columnas_a_encolumnar = ['Cantidad total de VIVIENDAS visitadas',

                          'Cantidad de inmuebles INSPECCIONADOS',

                          'Total de inmuebles positivos',

                          'Recipientes POSITIVOS',

                          'Recipientes NEGATIVOS']
```

```python
# Encolumnamos las columnas seleccionadas en un nuevo DataFrame y llamamos a la columna "Cantidad"

VIGILANCIA_encolumnada = pd.concat([VIGILANCIA[col].astype(str) for col in columnas_a_encolumnar],

                                    ignore_index=True)
VIGILANCIA_encolumnada = VIGILANCIA_encolumnada.to_frame(name='Cantidad')
```

```python
ID_localidad_X5 = ['ID_Localidad',

                          'ID_Localidad',

                          'ID_Localidad',

                          'ID_Localidad',

                          'ID_Localidad']
ID_localidad_X5 = pd.concat([VIGILANCIA[col].astype(str) for col in ID_localidad_X5],

                                    ignore_index=True)
ID_localidad_X5 = ID_localidad_X5.to_frame(name='ID_Localidad')
```

```python
Fecha_de_la_actividad_X5 = ['Fecha de la actividad',

                          'Fecha de la actividad',

                          'Fecha de la actividad',

                          'Fecha de la actividad',

                          'Fecha de la actividad']
Fecha_de_la_actividad_X5 = pd.concat([VIGILANCIA[col].astype(str) for col in Fecha_de_la_actividad_X5],

                                    ignore_index=True)
Fecha_de_la_actividad_X5 = Fecha_de_la_actividad_X5.to_frame(name='Fecha de la actividad')
```

```python
Responsables_de_las_actividades_X5 = ['Responsables de las actividades',

                          'Responsables de las actividades',

                          'Responsables de las actividades',

                          'Responsables de las actividades',

                          'Responsables de las actividades']
Responsables_de_las_actividades_X5 = pd.concat([VIGILANCIA[col].astype(str) for col in Responsables_de_las_actividades_X5],

                                    ignore_index=True)
Responsables_de_las_actividades_X5 = Responsables_de_las_actividades_X5.to_frame(name='Responsables de las actividades')
```

```python
Provincia_X5 = ['Provincia',

                          'Provincia',

                          'Provincia',

                          'Provincia',

                          'Provincia']
Provincia_X5 = pd.concat([VIGILANCIA[col].astype(str) for col in Provincia_X5], ignore_index=True)
Provincia_X5 = Provincia_X5.to_frame(name='Provincia')
```

```python
Departamento_X5 = ['Departamento',

                    'Departamento',

                    'Departamento',

                    'Departamento',

                    'Departamento']
Departamento_X5 = pd.concat([VIGILANCIA[col].astype(str) for col in Departamento_X5], ignore_index=True)
Departamento_X5 = Departamento_X5.to_frame(name='Departamento')
```

```python
Control_focal_X5 = ['Control focal',

                    'Control focal',

                    'Control focal',

                    'Control focal',

                    'Control focal']
Control_focal_X5 = pd.concat([VIGILANCIA[col].astype(str) for col in Control_focal_X5], ignore_index=True)
Control_focal_X5 = Control_focal_X5.to_frame(name='Control focal')
```

```python
#Creamos la columna TIPO repitiendo las columnas seleccionadas de acuerdo al numero de filas
TIPO = [item for item in ['Viviendas visitadas', 'Inmuebles inspeccionados', 'Inmuebles positivos', 'Recipientes positivos', 'Recipientes negativos'] for _ in range(Filas)]
print(TIPO)
```
